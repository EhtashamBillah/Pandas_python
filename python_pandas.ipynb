#### Agrregation based on one column

```python
# reset_index returns the output as dataframe. we can also write [["year"]] (as a list) to return the output as dataframe
df.groupby("state")["year"].count().reset_index()
```

```python
df.groupby("state")[["number"]].median()
```

```python
# 3rd quartile or 75th percentile. 
# A quartile is the value of the quantile at the probabilities 0.25, 0.5 and 0.75.
# this table can also be obtained using lambda function as in next example
df.groupby("state")["number"].quantile(.75)
```

```python
# 3rd quartile/ 75th percentile of number for each state
import numpy as np
df.groupby("state")["number"].apply(lambda x : np.percentile(x,75)).reset_index()
```

#### Aggregation based on two columns

```python
group_by_two_col = df.groupby(["month", "state"])["number"].mean().reset_index()
print(group_by_two_col)
```

```python
# pivoting the table above
group_by_two_col.pivot(index = "state",
                      columns = "month",
                      values = "number")
```

```python
# exact same table as above can also be obtained as
df.pivot_table(index = "state",
               columns = "month",
               values = "number",
               aggfunc= "mean")
```

#### creating column based on another column

```python
# creating a variable named "day" from the date column using lambda expression
find_day = lambda x: x.split("-")[-1]
df["day"] = df["date"].apply(find_day)
```

```python
find_mill = lambda x : "previous_millennium" if int(x.split("-")[0]) < 2000 else "new_millennium"
df["millenium"] = df["date"].apply(find_mill)
```

```python
df.head()
```

```python
# using map function to create new variable..very useful to create dummy variable
df["month_digit"] = df["month"].map(    
    
    {'Janeiro' : 1,
     'Fevereiro' : 2,
     'Março' : 3,
     'Abril' : 4,
     'Maio' : 5,
     'Junho' : 6,
     'Julho' : 7,
     'Agosto' : 8,
     'Setembro' : 9,
     'Outubro' : 10,
     'Novembro' : 11,
     'Dezembro' : 12}
)
```
